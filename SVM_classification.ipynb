```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

# Step 1: Load the dataset
df = pd.read_csv("/content/drive/MyDrive/DATASET/wine.csv")

# Step 2: Separate features and target
X = df.drop(columns=['Wine'])  # Features
y = df['Wine']                 # Target class

# Step 3: Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 4: Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# Step 5: Initialize and train SVM classifier
svm_model = SVC(kernel='rbf', C=1.0, gamma='scale')  # RBF kernel is good for nonlinear data
svm_model.fit(X_train, y_train)

# Step 6: Predict and evaluate
y_pred = svm_model.predict(X_test)

# Step 7: Print evaluation results
print("✅ Accuracy:", accuracy_score(y_test, y_pred))
print("\n📊 Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\n📋 Classification Report:\n", classification_report(y_test, y_pred))
```
